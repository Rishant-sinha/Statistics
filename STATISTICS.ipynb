#5.  Discuss the role of random sampling in making inferences about populations.

Role of Random Sampling in Making Inferences About Populations
Random sampling is a fundamental statistical technique used to make inferences about a population based on a subset (sample). It ensures that every member of the population has an equal chance of being selected, reducing bias and increasing the reliability of conclusions.

Why Random Sampling is Important
Ensures Representativeness

A well-drawn random sample reflects the characteristics of the entire population, allowing for generalization.

Example: If a university wants to study student satisfaction, randomly selecting students from different majors ensures a balanced perspective.

Reduces Bias

Avoids systematic errors that occur when certain groups are overrepresented or underrepresented.

Example: If only volunteers participate in a survey, results may be skewed toward more opinionated individuals.

Allows for Statistical Inference

By applying probability theory, conclusions drawn from a sample can be extended to the entire population using measures like confidence intervals and hypothesis testing.

Example: If a survey finds that 60% of sampled voters favor a candidate, statistical techniques can estimate the true percentage in the entire electorate.

Enables Error Measurement

Sampling comes with some uncertainty, but methods like margin of error and standard error quantify this uncertainty, making results more reliable.

Types of Random Sampling
Simple Random Sampling (SRS) â€“ Every individual has an equal chance of selection.

Example: Drawing names from a hat.

Stratified Random Sampling â€“ Population is divided into subgroups (strata), and random samples are taken from each.

Example: Sampling an equal number of men and women in a survey.

Systematic Sampling â€“ Selecting every nth individual from a list.

Example: Picking every 10th customer entering a store.

Cluster Sampling â€“ The population is divided into clusters, and entire clusters are randomly selected.

Example: Selecting random school districts instead of individual students.

#6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Skewness measures the asymmetry of a probability distribution relative to its mean. It indicates whether data points are more concentrated on one side of the distribution, affecting the interpretation of central tendency.

Types of Skewness
Positive Skew (Right-Skewed Distribution)

The right tail is longer, meaning extreme values are concentrated on the higher end.

Mean > Median > Mode

Example: Income distribution (few people earn extremely high salaries while most earn moderate amounts).

Negative Skew (Left-Skewed Distribution)

The left tail is longer, meaning extreme values are concentrated on the lower end.

Mean < Median < Mode

Example: Test scores where most students score high but a few score very low.

Zero Skew (Symmetric Distribution)

The distribution is balanced with equal tails on both sides.

Mean = Median = Mode

Example: Heights of adults in a population (assuming normal distribution).

How Skewness Affects Data Interpretation
Impact on Measures of Central Tendency

In skewed data, the mean is pulled toward the longer tail, making the median a better measure of central tendency.

Example: In right-skewed salary data, the median salary is more representative than the mean salary.

Effect on Data Analysis

Right-skewed data may indicate outliers (e.g., extreme income levels).

Left-skewed data may indicate minimum limits (e.g., exam grading with a passing cutoff).

Influence on Statistical Methods

Many statistical tests assume normality (zero skew). If data is skewed, data transformations (e.g., log transformation) may be needed for accurate analysis.
